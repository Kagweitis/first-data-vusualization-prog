```python
import pandas as pd
from matplotlib import pyplot as plt

x=[3,4,5]
y=[4,6,7]
z=[10,12,34]
plt.plot(x,y)
plt.plot(x,z)
plt.title("test plot")
plt.xlabel("x")
plt.ylabel("y and z")
plt.legend(["this is y", "this is z"])
plt.show()
```

```python
data=pd.read_csv('countries.csv')
```

```python
data
```

```python
#compare pop growth between U.S and china
us=data[data.country=='United States']
```

```python
china=data[data.country=='China']
```

```python
us
china
```

```python
plt.plot(us.year, us.population/10**6) 
plt.plot(china.year, china.population/10**6)
plt.legend(["this is the US", "this is China"])
plt.xlabel('population')
plt.ylabel('years')
plt.show()
```

```python
#to find percentage growth in pop
us.population
```

```python
us.population /us.population.iloc[0]* 100 #iloc is used to select rows and columns by number
```

```python
plt.plot(us.year, us.population /us.population.iloc[0]* 100) 
plt.plot(china.year, china.population/china.population.iloc[0]*100)
plt.legend(["this is the US", "this is China"])
plt.xlabel('population growth')
plt.ylabel('years')
plt.show()
```
